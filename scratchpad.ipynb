```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Define the  Quadratic Function
def quadratic(x,a=-1,b=0,c=10):
  return a*x**2 + b*x + c
```

```
#Define the WaterFall Model
stages = ["Requirements","Design","Development","Testing","Deployment"]
x_positions=np.linspace(-2, 2, len(stages)) #Equally spaced x positions
y_positions=quadratic(x_positions) #Compute y positions using thw quadratic function
#Plot the Quadratic curve
x_curve=np.linspace(-2.5,2.5,500)
y_curve=quadratic(x_curve)
plt.figure(figsize=(10,6))
plt.plot(x_curve,y_curve,color="blue",linestyle="--",label="Quadratic curve")
#Add markers and connect them with vertical lines to create the WaterFall effect
for i in range(len(stages)):
  if i>0:
    #Draw a vertical drop to Simulate a waterfall
    plt.plot(
        [x_positions[i-1],x_positions[i]],
        [y_positions[i-1],y_positions[i-1]],
        color="red",
        linewidth=2
    )
    plt.plot(
        [x_positions[i],x_positions[i]],
        [y_positions[i-1],y_positions[i]],
        color="blue",
        linewidth=2
    )
  # Place markers and labels for all stages
  plt.scatter(x_positions[i], y_positions[i], color='red', zorder=5)
  plt.text(x_positions[i], y_positions[i], stages[i], ha='center', va='center', fontsize=10)
# Style the plot
plt.title("Waterfall Model Representation using Quadratic Curve", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Add horizontal and vertical reference lines
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

# Customize the grid
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)

# Add a legend (ensure you add labels in the plot commands for the legend to work)
plt.legend()

# Display the plot
plt.show()

```
